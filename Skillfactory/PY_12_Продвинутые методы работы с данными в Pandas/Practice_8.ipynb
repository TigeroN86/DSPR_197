```python
import pandas as pd

movies = pd.read_csv('data/ratings_movies.csv')
movies.head()
```

```python
#библиотека для регулярных выражений
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
movies['year_release'] = movies['title'].apply(get_year_release)
movies['year_release'].info()
```

```python
movies[movies['year_release'] == 1999].groupby('title')['rating'].mean().sort_values()
```

```python
movies[movies['year_release'] == 2010].groupby('genres')['rating'].mean().sort_values()
```

```python
movies.groupby('userId')['genres'].nunique().sort_values(ascending=False)
```

```python
movies.groupby('userId')['rating'].agg(['count', 'mean']).sort_values(by=['count', 'mean'], ascending=[True, False])
```

```python
movies[movies['year_release'] == 2018].groupby('genres')['rating'].agg(['mean', 'count']).sort_values(['mean', 'count'], ascending=[False, False])
```
